# **Get Average Flow**

```python
import pickle
def load_pickle_file(file_name):
  with open(file_name, 'rb') as handle:
      return pickle.load(handle)
```

```python
flow_values = load_pickle_file("flow_dict.pickle")
average_flow_value = dict()
import numpy as np
for key in flow_values:
  values = flow_values[key]
  average = np.average(values)
  average_flow_value[key] = average
```
